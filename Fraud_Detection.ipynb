```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 100000

customer_id=np.random.randint(1000,5000,n)

amount=np.random.randint(50,5000,n)

city=np.random.choice(
[
"Delhi",
"Mumbai",
"Bangalore",
"Chennai",
"Hyderabad"
],n)

ip=np.random.choice(
[
"192.168.1.1",
"172.16.5.4",
"10.0.0.5",
"45.67.12.5",
"100.45.21.3"
],n)

merchant=np.random.choice(
[
"Amazon",
"Flipkart",
"Myntra",
"Swiggy",
"Zomato"
],n)

transactions=pd.DataFrame({
"Customer_ID":customer_id,
"Amount":amount,
"City":city,
"IP_Address":ip,
"Merchant":merchant
})

transactions.to_csv("transactions.csv",index=False)

transactions.head()
```

```python
import pandas as pd

df=pd.read_csv("transactions.csv")

df.head()
```
